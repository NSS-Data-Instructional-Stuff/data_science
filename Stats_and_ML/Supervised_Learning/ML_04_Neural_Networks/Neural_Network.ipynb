```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def generateData(size = 50, sigma = 0.2, random_state = None):
    if random_state:
        np.random.seed(random_state)
    
    x1 = np.random.uniform(low = 0, high = 1, size = size)
    x2 = np.random.uniform(low = 0, high = 1, size = size)
    y = np.sin(2*np.pi*x1) + np.cos(2*np.pi*x2) + np.random.normal(loc = 0, scale = sigma, size = size)
    
    X = np.c_[x1, x2]
    
    return X, y
```

```python
X, y = generateData(size = 50, random_state = 321)
X_eval, y_eval = generateData(size = 10000)
```

```python
plt.figure(figsize = (10,8))

u = np.linspace(0, 1, 100)
v = np.linspace(0, 1, 100)
U,V = np.meshgrid(u, v)
Z = np.sin(2*np.pi*U) + np.cos(2*np.pi*V)
plt.pcolor(U, V, Z, vmin = -2, vmax = 2, shading = 'nearest')
plt.colorbar()
plt.title('True Values', fontsize = 24, fontweight = 'bold')
cm = plt.get_cmap();
```

First, we'll build a model that takes our data and maps it into a 5-dimensional space.

```python
model = Sequential()

model.add(Dense(5, input_dim = X.shape[1], activation = 'tanh'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.fit(X, y, epochs = 5000)
```

```python
mean_squared_error(y, model.predict(X))
```

```python
mean_squared_error(y_eval, model.predict(X_eval))
```

```python
Z = model.predict(np.array([U.ravel(), V.ravel()]).T)
Z = Z.reshape(U.shape)

fig, ax = plt.subplots(figsize = (10,8))

plt.pcolor(U, V, Z, vmin = -2, vmax = 2, shading = 'nearest')
plt.colorbar()

cm = plt.get_cmap()
plt.scatter(X[:,0], X[:,1], edgecolor = 'white', linewidth = 2,
            s = 100, c = y, cmap = cm, zorder = 100);
```

Maybe mapping into a 10-dimensional space would do better.

```python
model = Sequential()

model.add(Dense(10, input_dim = X.shape[1], activation = 'tanh'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
model.fit(X, y, epochs = 5000)
```

```python
mean_squared_error(y, model.predict(X))
```

```python
mean_squared_error(y_eval, model.predict(X_eval))
```

```python
Z = model.predict(np.array([U.ravel(), V.ravel()]).T)
Z = Z.reshape(U.shape)

fig, ax = plt.subplots(figsize = (10,8))

plt.pcolor(U, V, Z, vmin = -2, vmax = 2, shading = 'nearest')
plt.colorbar()

cm = plt.get_cmap()
plt.scatter(X[:,0], X[:,1], edgecolor = 'white', linewidth = 2,
            s = 100, c = y, cmap = cm, zorder = 100);
```

We didn't see much of an improvement. Let's see what happens if we add another layer.

```python
model = Sequential()

model.add(Dense(10, input_dim = X.shape[1], activation = 'tanh'))
model.add(Dense(10, activation = 'tanh'))
model.add(Dense(1, activation = 'linear'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam', )
```

```python
model.fit(X, y, epochs = 5000)
```

```python
mean_squared_error(y_eval, model.predict(X_eval))
```

For reference, the irreducible error in this data generation process is equal to 0.2^2 = 0.04.

```python
Z = model.predict(np.array([U.ravel(), V.ravel()]).T)
Z = Z.reshape(U.shape)

fig, ax = plt.subplots(figsize = (10,8))

plt.pcolor(U, V, Z, vmin = -2, vmax = 2, shading = 'nearest')
plt.colorbar()

cm = plt.get_cmap()
plt.scatter(X[:,0], X[:,1], edgecolor = 'white', linewidth = 2,
            s = 100, c = y, cmap = cm, zorder = 100);
```
